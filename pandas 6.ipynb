```python
dfA = pd.DataFrame({'Key':['A','B','C','D'],
                   'A':[1,2,3,5]})
dfB = pd.DataFrame({'Key':['E','F','G','H'],
                   'B':[6,7,8,9]})
pd.merge(dfA, dfB, on='Key')
```

```python
dfA = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
dfB = pd.DataFrame({'Key':['E','F','G','H'],
                   'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
pd.merge(dfA, dfB, on='Key')
```

```python
dfA = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
dfB = pd.DataFrame({'Key':['E','F','G','H'],
                   'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
pd.merge(dfA, dfB, how= 'right', on='Key')
```

```python
dfA = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
dfB = pd.DataFrame({'Key':['E','F','G','H'],
                   'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
pd.merge(dfA, dfB, how= 'inner', on='Key')
```

```python
dfA = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
dfB = pd.DataFrame({'Key':['E','F','G','H'],
                   'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
pd.merge(dfA, dfB, how= 'outer', on='Key')
```

# joining

```python
import pandas as pd
left = pd.DataFrame({'A':['K0','K1','K2','K3'],
                     'B':['A0','A1','A2','A3']},
                     index =['B0','B1','B2','B3'])

left
```

```python
right = pd.DataFrame({'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3']},
                      index = ['B0','B1','B2','B3'])

right
```

```python
left.join(right, how = 'left',lsuffix = '_left',rsuffix = '_right')
```

```python
left.join(right, how = 'outer',lsuffix = '_left',rsuffix = '_right')
```

```python
left.join(right)
```

```python
left.join(right, how= 'outer')
```
